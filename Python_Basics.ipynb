Q1. What is Python, and why is it popular?

Python is a high-level programming language known for its simplicity and readability. It’s popular because it’s beginner-friendly, has tons of libraries, and is used in almost everything — from web development to AI and data science.

Q2. What is an interpreter in Python?

An interpreter reads and executes Python code line by line instead of compiling it all at once. That’s why you can quickly test and debug Python programs.

Q3. What are pre-defined keywords in Python?

Pre-defined keywords are special reserved words that have specific meanings in Python, like if, for, while, True, def, etc. You can’t use them as variable names.

Q4. Can keywords be used as variable names?

No, keywords can’t be used as variable names because they already have a defined meaning in Python. Doing so would confuse the interpreter.

Q5. What is mutability in Python?

Mutability means whether an object can be changed after it’s created. Mutable objects (like lists) can be modified, while immutable ones (like tuples) can’t.

Q6. Why are lists mutable, but tuples are immutable?

Lists are mutable because you can add, remove, or change elements in them. Tuples are immutable to keep data safe and fixed — once created, they can’t be altered.

Q7. What is the difference between “==” and “is” operators in Python?

== checks if the values of two objects are the same, while is checks if they refer to the same memory location.

Q8. What are logical operators in Python?

Logical operators are used to combine conditional statements. They are and, or, and not.

Q9. What is type casting in Python?

Type casting means converting a value from one data type to another, like turning an integer into a string or a float into an integer.

Q10. What is the difference between implicit and explicit type casting?

In implicit casting, Python automatically converts data types, while in explicit casting, you do it manually using functions like int(), float(), or str().

Q11. What is the purpose of conditional statements in Python?

Conditional statements let your program make decisions and run certain parts of the code only if a condition is true.

Q12. How does the elif statement work?

The elif statement checks another condition if the previous if statement is false. It’s like saying “if not this, then check that.”

Q13. What is the difference between for and while loops?

A for loop runs a fixed number of times, usually through a sequence, while a while loop runs until a condition becomes false.

Q14. Describe a scenario where a while loop is more suitable than a for loop.

A while loop is better when you don’t know how many times the loop should run — like asking a user for input until they type “exit.”

```python
# Q4. Write a program that checks if a given word is a Python keyword.

import keyword
word = input("Enter a word: ")
if keyword.iskeyword(word):
    print(word, "is a Python keyword.")
else:
    print(word, "is not a Python keyword.")
```

```python
# Q5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each.

my_list = [1, 2, 3]
my_tuple = (1, 2, 3)

my_list[0] = 10  # Works fine
print("Modified list:", my_list)

try:
    my_tuple[0] = 10  # Will cause an error
except TypeError:
    print("Tuples can’t be changed!")
```

```python
# Q6. Write a function to demonstrate the behavior of mutable and immutable arguments.

def modify_data(a_list, a_number):
    a_list.append(10)
    a_number += 5
    print("Inside function:", a_list, a_number)

lst = [1, 2, 3]
num = 7
modify_data(lst, num)
print("Outside function:", lst, num)
```

```python
# Q13. Write a Python program to find the sum of all even numbers between 1 and 50.

total = 0
for i in range(1, 51):
    if i % 2 == 0:
        total += i
print("Sum of even numbers:", total)
```

```python
# Q14. Write a program to reverse a string using a while loop.

s = input("Enter a string: ")
rev = ""
i = len(s) - 1
while i >= 0:
    rev += s[i]
    i -= 1
print("Reversed string:", rev)
```

```python
# Q15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

num = int(input("Enter a number: "))
fact = 1
i = 1
while i <= num:
    fact *= i
    i += 1
print("Factorial:", fact)
```
